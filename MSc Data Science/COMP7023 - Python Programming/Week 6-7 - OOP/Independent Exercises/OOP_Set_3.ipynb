## Q1 — BoundedList

**Create a class `BoundedList` that stores items but never allows the list to exceed a fixed size.**
Methods:
- `add(x)` → adds to end unless full.
- `pop()` → removes and returns last element.
- `is_full()` / `is_empty()`.
Internal list must never exceed the capacity.

```python
# Your code here
class BoundedList():
    def __init__(self, bound):
        self.list = []
        self.bound = bound

    def add(self, x):
        if len(self.list) < self.bound:
            self.list.append(x)        

    def pop(self):
        if len(self.list) != 0:
            return self.list.pop()

    def is_full(self):
        return len(self.list) == self.bound

    def is_empty(self):
        return len(self.list) == 0
```

```python
# Tests
bl = BoundedList(3)
assert bl.is_empty()
bl.add(1); bl.add(2); bl.add(3)
assert bl.is_full()
bl.add(99)  # ignored
assert bl.pop()==3
assert not bl.is_full()
print("Q1 OK")
```

**Create a class `ScoreTracker` that maps names → scores.**
Requirements:
- `add(name, score)` adds or updates.

- `average()` returns mean score.
- Internal dict must not leak.

```python
# Your code here
class ScoreTracker():
    def __init__(self):
        self.dict_q2 = {}

    def add(self, name, score):
        self.dict_q2[name] = score
        print(f"Dict = {self.dict_q2}")
    
    def highest(self):
        name = max(self.dict_q2, key=lambda n: self.dict_q2[n])
        return (name, self.dict_q2[name])
    
    def average(self):
        total = 0
        for i in self.dict_q2:
            total += self.dict_q2[i]
        return total/len(self.dict_q2)
```

## Q3 — WordBag

**Create a class `WordBag` storing word counts.**
- `add(word)` increases count.
- `remove(word)` decreases count or deletes.
- `count(word)` returns count.
- `unique()` returns number of distinct words.

```python
# Your code here
class WordBag():
    def __init__(self):
        self.bag = {}

    def add(self, word):
        if word not in self.bag:
            self.bag[word] = 1
        else:
            self.bag[word] += 1
    
    def remove(self, word):
        if self.bag[word] == 1:
            del self.bag[word]
        else:
            self.bag[word] -= 1
    
    def count(self, word):
        return self.bag[word]

    def unique(self):
        return len(self.bag)
```

```python
# Tests
wb=WordBag()
wb.add("hi"); wb.add("hi"); wb.add("bye")
assert wb.count("hi")==2
wb.remove("hi")
assert wb.count("hi")==1
assert wb.unique()==2
print("Q3 OK")
```

## Q4 — StudentRegistry

**Create a class storing student records as dicts: {name, age}.**
- `add_student(name, age)`
- `find(name)` → returns copy of record or None.
- `all_ages()` → list of ages.
- Must store internal copies.

```python
# Your code here
class StudentRegistry():
    def __init__(self):
        self.reg = {}

    def add_student(self, name, age):
        self.reg[name] = {"name": name, "age": age}

    def find(self, name):
        if name in self.reg:
            return self.reg[name].copy()
    
    def all_ages(self):
        ages_list = []
        for student in self.reg:
            ages_list.append(self.reg[student]["age"])
        return ages_list
```

```python
# Tests
sr=StudentRegistry()
sr.add_student("A",20); sr.add_student("B",25)
rec=sr.find("A")
assert rec["age"]==20
rec["age"]=99
assert sr.find("A")["age"]==20
assert sorted(sr.all_ages())==[20,25]
print("Q4 OK")
```

## Q5 — SimpleTable

**Create a Table class:**
- Stores rows as dicts.
- `add_row(d)` stores COPY.
- `select(col)` returns list of that column’s values.
- `filter(col,val)` returns list of row copies where `row[col]==val`.

```python
# Your code here
class SimpleTable():
    def __init__(self):
        self.rows = []
    
    def add_row(self, d):
        self.rows.append(d.copy())

    def select(self, col):
        column_list = []
        for row in self.rows:
            column_list.append(row[col])
        return column_list
    
    def filter(self, col, val):
        filtered_list = []
        for row in self.rows:
            if row[col]==val:
                filtered_list.append(row.copy())
        return filtered_list
```

```python
# Tests
t=SimpleTable()
t.add_row({"A":1,"B":2})
t.add_row({"A":1,"B":9})
assert t.select("A")==[1,1]
rows=t.filter("B",2)
assert rows[0]["A"]==1
rows[0]["A"]=99
assert t.filter("B",2)[0]["A"]==1
print("Q5 OK")
```

## Q6 — Rectangle

**A simple geometry class:**
- init with width, height.
- `area()`, `perimeter()`
- `scale(f)` returns NEW Rectangle scaled.
Ensure immutability: scaling must not change the original.

```python
# Your code here
class Rectangle():
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width*self.height
    
    def perimeter(self):
        return (2*self.width) + (2*self.height)

    def scale(self, f):
        new_w = self.width*2
        new_h = self.height*2
        return Rectangle(new_w, new_h)
```

```python
# Tests
r=Rectangle(2,4)
assert r.area()==8
assert r.perimeter()==12
r2=r.scale(2)
assert r2.area()==32
assert r.area()==8
print("Q6 OK")
```

## Q7 — HistoryList

**Like a list but keeps a history of all values ever appended.**
- `add(x)`
- `current()` returns current list copy.
- `history()` returns list of snapshots (each snapshot is its own list copy).

```python
# Your code here
class HistoryList():
    def __init__(self):
        self.history_list = []
        self.history_list_snapshot = []

    def add(self, value):
        self.history_list .append(value)
        self.history_list_snapshot.append(self.history_list .copy())


    def current(self):
        return self.history_list .copy()
    
    def history(self):
        return [snap.copy() for snap in self.history_list_snapshot]
```

```python
# Tests
h=HistoryList()
h.add(1); h.add(2)
cur=h.current()
assert cur==[1,2]
hist=h.history()
assert len(hist)==2
hist[0].append(99)
assert h.history()[0]==[1]
print("Q7 OK")
```
