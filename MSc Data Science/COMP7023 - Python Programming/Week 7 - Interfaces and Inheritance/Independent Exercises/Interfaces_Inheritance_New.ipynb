## Question 1
Define a class `Vehicle` with `brand` and method `info()`.
Define subclass `Car` adding `doors` and overriding `info()`.
Instantiate `Car("BMW", 3)` and call `info()`. 

````python
# your code here
class Vehicle():
    def __init__(self, brand):
        self.brand = brand
    
    def info(self):
        pass

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)
        self.doors = doors
    
    def info(self):
        return f"{self.brand}, {self.doors}"
````

````python
# asserts
car = Car("BMW", 3)
assert car.brand == "BMW"
assert car.doors == 3
out = car.info()
assert "BMW" in out
assert "3" in out
````

## Question 2
Override `deposit` in `RewardsAccount` using `super()` to add a reward after depositing.

````python
# your code here
class Account:
    """
    A basic bank account.
    """
    def __init__(self, holder):
        self.holder = holder
        self.balance = 0

    def deposit(self, amount):
        """
        Increase the balance by amount and return the new balance.
        """
        self.balance += amount
        return self.balance

class RewardsAccount(Account):
    def deposit(self, amount):
        super().deposit(amount)
        self.balance += 1
````

````python
# asserts
a = RewardsAccount("A")
a.deposit(10)
assert a.balance == 11
````
