**Add a method `age()` that calculates how old the car is.**
- Use a fixed year (e.g., 2025).
- Return or print the car's age.

```python
# Your code here
class Car():
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year    
        return None
    
    def drive(self):
        print("Driving...")

    def age(self):
        age = 2025 - self.year
        return age

car1 = Car("BMW", 2023)
print(car1.age())
```

## Question 4
**Create a class `Book` that stores `title`, `author`, and `price`.**
- Add a method `discount(percent)` that reduces the price.
- Create one object and test it.

```python
# Your code here
class Book():
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    
    def discount(self, percent):
        return self.price * (((100-percent))/100)

book1 = Book("Title1", "Author 1", 10)
print(f"Discounted price = £{int(book1.discount(10))}")
```

## Question 9
**Add an `add()` method to your `Inventory` class.**
- It should take `name`, `qty`, and `price`.
- If the item exists, update quantity.
- If not, add it.

```python
# Your code here
class Inventory():
    def __init__(self):
        self.items = {}

    def add(self, name, qty, price):
        self.items[name] = {"qty": qty, "price":price}

inv = Inventory()
inv.add("Apple", 12, 0.30)
inv.add("Banana", 15, 0.10)

for i in inv.items:
    print(f"Item: {i}")
    print(f"Quantity = {inv.items[i]["qty"]}")
    print(f"Price = ${inv.items[i]["price"]:.2f}")
    print()
```

## Question 10
**Add a method `total()` to compute total value of all inventory items.**

```python
# Your code here
class Inventory():
    def __init__(self):
        self.items = {}

    def add(self, name, qty, price):
        self.items[name] = {"qty": qty, "price":price}
    def total(self):
        return len(self.items)

inv = Inventory()
inv.add("Apple", 12, 0.30)
inv.add("Banana", 15, 0.10)

print(inv.total())

inv.add("Raspberry", 30, 0.02)

print(inv.total())
```
