```python
# Question 1
def any_char(func, string):
    i = 0
    while i < len(string):
        if func(string[i]):
            return True
        i += 1
    return False
```

```python
# Question 2
def filter_str(func, string):
    result = ""
    i = 0
    while i < len(string):
        if func(string[i]) == True:
            result += string[i]
        i += 1
    return result
```

```python
# Question 3
def map_str(func, string):
    """Recursive function"""
    result = ""
    for i in string:
        result += func(i)
    return result
```

```python
# Question 4
def filter_str(func, string):
    """Recursive function"""
    return map_str(lambda c: c if func(c) else "", string)
```

```python
# Question 5
def any_char_strict(func, string):
    """New any_char function"""
    mapped = map_str(lambda c: "1" if func(c) else "", string)
    return mapped != ""
```

```python
# Question 6
def apply_until(f, p , x):
    counter = 0
    while not p(x):
        x = f(x)
        counter += 1
    return counter
```

```python
# Question 7
def find_first(func, string):
    """Docstring"""
    i = 0
    while i < len(string):
        if func(string[i]):
            return i
        i += 1
    return -1
```

```python
# Question 8:
def compose1(f, g):
    return lambda x: f(g(x))

def compose_repeated(n, f):
    result = lambda x: x
    i = 0
    while i < n:
        result = compose1(f, result)
        i += 1
    return result
```
