```python
# camelCase - camel.py
camel_name = input("Enter variable name").strip()

name_list = list(camel_name)
for i in range(len(name_list)):
    if name_list[i].isupper():
        name_list[i] = name_list[i].lower()
        name_list.insert(i,"_")

snake_name = ""
snake_name = snake_name.join(name_list)
print(snake_name)
```

```python
# Better way of the above
camel_name = input("Enter variable name: ").strip()
snake_name = ""

for ch in camel_name:
    if ch.isupper():
        snake_name += "_" + ch.lower()
    else:
        snake_name += ch

print(snake_name)
```
