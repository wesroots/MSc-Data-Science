```python
# File Extensions - extensions.py
file_name = input("File name: ").strip().lower()
file_name_list = list(file_name)
index = 0
for i in file_name_list:
    if i == ".":
        extension_name = ""
        extension_name = extension_name.join(file_name_list[index:])
        print(extension_name)
        break
    else:
        index += 1
```

```python
# Math Interpreter - interpreter.py
raw_input = input("Equation")
list1 = raw_input.split(" ")

x = float(list1[0])
y = str(list1[1])
z = float(list1[2])

if y == "+":
    result = x+z
    print(f"{result:.1f}")
elif y == "-":
    result = x-z
    print(f"{result:.1f}")
elif y == "*":
    result = x*z
    print(f"{result:.1f}")
else:
    result = x/z
    print(f"{result:.1f}")
```

```python
# Meal Time - meal.py
time = input("Enter time: ").strip().split(":")
time_int = []

for i in time:
    time_int.append(int(i))

hour = time_int[0]
minutes = time_int[1]

if hour == 7:
    print(f"Breakfast time - it's {hour}:{minutes}")
elif hour == 12:
    print(f"Lunch time - it's {hour}:{minutes}")
elif hour == 18:
    print(f"Dinner time - it's {hour}:{minutes}")
else:
    print(f"Not meal time - it's {hour}:{minutes}")
```

```python
# Proper modular design for the above
def main():
    time = input("What time is it?").strip()
    time = convert(time)
    if time >= 7 and time <8:
        print("Breakfast time")
    elif time >= 12 and time < 13:
        print("Lunch time")
    elif time >= 18 and time <19:
        print("Dinner time")

def convert(time):
    time = time.split(":")
    for i in range(len(time)):
        time[i] = float(time[i])
    time = time[0] + round(time[1]/60,2)
    return time
    

if __name__ == "__main__":
    main()
```
