# Lists

## Intro

```python
# Square brackets indicate a list
bicycles = ["trek", "cannondale", "redline", "specialised"]
print(bicycles)
```

```python
# Accessing the first item - index starts at 0
print(bicycles[0].title())

# Accessing the last item
print(bicycles[-1].title())

# Accessing second last
print(f"Second last: {bicycles[-2].title()}")
```

## Modifying Elements

```python
# Modifying elements in a list
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

motorcycles[0] = "ducati"
print(motorcycles)
```

```python
# Appending elements to the end of a list
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

motorcycles.append("ducati")
print(motorcycles)
```

```python
# Inseting elements in a list
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

motorcycles.insert(0, "ducati")
print(motorcycles)
```

## Removing Elements

```python
# Removing elements from a list
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

del motorcycles[1]
print(f"{motorcycles} - yamaha is gone")

# Removing by value
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

motorcycles.remove("honda")
print(f"{motorcycles} - honda is gone")

# Removing by value removes the first instance
```

```python
# Removing item using pop() method
motorcycles = ["honda", "yamaha", "suzuki"]
print(motorcycles)

popped_motorcycle = motorcycles.pop(1)
print(motorcycles)
print(popped_motorcycle)

# If no index returned pop() removes last element
```

## Sorting a List

```python
# Sorting a list permanently with sort()
cars = ["bmw", "audi", "toyota", "subaru"]
cars.sort()
print(cars)

# Sorting in reverse
cars = ["bmw", "audi", "toyota", "subaru"]
cars.sort(reverse=True)
print(cars)
```

```python
# Sorting a list temoporarily with sorted() function
cars = ["bmw", "audi", "toyota", "subaru"]
print("Here is the original list:")
print(cars)

print("\nHere is the sorted list:")
print(sorted(cars))

print("\nHere is the original list again:")
print(cars)
```

```python
# Pringing in reverse order (not alphabetically)
cars = ["bmw", "audi", "toyota", "subaru"]
print(cars)

cars.reverse()
print(cars)
```

```python
# Finding length
cars = ["bmw", "audi", "toyota", "subaru"]
print(len(cars))
```
