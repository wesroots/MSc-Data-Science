# Working with Lists

### Looping

```python
# Looping through a list
magicians = ["alice", "david", "carolina"]
counter = 1
for magician in magicians:
    print(f"{magician} - magician number {counter}")
    counter +=1 
```

```python
# Range() function to loop through numbers 1 to 5
for value in range(1,6):
    print(f"Number {value}")

# Range() can create a list of numbers
numbers = list(range(1,6))
print(numbers)
```

### Simple Maths

```python
# Simple stats with lists
digits = list(range(1,10))
digits.append(0)
print(digits)

print(min(digits))
print(max(digits))
print(sum(digits))
```

### Working with part of a list

```python
# Slicing a list
players = ["charles", "martina", "michael", "florence", "eli"]
print(players[0:3])
print("(Prints from index 0 to 3)")

# Omitting the first index automatically starts slice at beginning of list
```

### Looping through a slice

```python
players=["charles", "martina", "michael", "florence", "eli"]
print("First three players:")
for player in players[:3]: # Up to but not including the third item
    print(player.title())
```

### Copying a List

```python
players_copy=players[:]
print(players)

# Adding to copied list
players_copy.append("extra player")
print(players_copy)
```

Using `list1 = list2` without a slice just creates another variable for the same list
