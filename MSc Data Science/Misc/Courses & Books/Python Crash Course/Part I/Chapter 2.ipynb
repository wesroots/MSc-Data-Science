```python
# Basic text formatting
first_name = "ada"
second_name = "lovelace"
full_name = f"{first_name.title()} {second_name.title()}"
print(full_name)
```

```python
# To add a tab to text use \t
print("Python")
print("\tPython")
```

```python
# To add a new line in string use \n
print("Languages: \n- Python \n- C \n- JavaScript")
```

```python
# Can be combined
print("Languages: \n\t Python \n\t C \n\t JavaScript")
```

```python
# Stripping white space
favourite_langauge = "python   "
favourite_langauge.rstrip().title() # Strips right side, lstrip() strips left side of white space
```

```python
# Making permenant changes
favourite_langauge = favourite_langauge.rstrip().title()
favourite_langauge
```

```python
# Removing prefixes
nostarch_url = "https://nostarch.com"
nostarch_url.removeprefix("https://")
```

```python
# Using apostrophe in string
message = "One of Python's strengths is its diverse community"
print(message)
```
