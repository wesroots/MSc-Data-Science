### 1. `Car` class

`__init__`: takes `make` and `year`.

Add an attribute `mileage` starting at 0.

Method `.drive(km)` → increase mileage by `km`.

Objects to create:

A 1997 BMW, drive it 100 km.

A 2020 Tesla, drive it 250 km.

```python
class car:
    def __init__(self, make, year):
        self.make = make
        self.year = year
        self.mileage = 0

    def drive(self,km):
        self.mileage += km
        return f"This car has driven {km}km"
    
e36 = car("BMW", 1997)

print(e36.drive(123))
```

### 2. `book` class

`__init__`: takes `title` and `author`.

Attribute: `is_checked_out = False` (default).

Method `.checkout()` → `set is_checked_out = True.`

Method `.return_book()` → `set is_checked_out = False.`

Objects to create:

A book 1984 by George Orwell. Checkout then return it.

A book Python Basics by Wes McKinney. Just checkout.

```python
class book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_checked_out = False

    def checkout(self):
        self.is_checked_out = True
        return f"{self.title} has been checked out"
    
    def returnbook(self):
        self.is_checked_out = False
        return f"{self.title} has been returned"

book1 = book("1984", "George Orwell")
book2 = book("Python Basics", "Wes McKinney")

print(book1.checkout())
print(book1.returnbook())

print(book2.checkout())
```

### 3. `calculator` class

`__init__`: no arguments needed.

Method `.add(a, b)` → return sum.

Method `.multiply(a, b)` → return product.

Objects to create:

A calculator that adds 2 + 3.

A calculator that multiplies 4 × 7.

```python
class calculator:
    def __init__(self):
        pass
    def add(self, *args):
        return sum(args)
    def multiply(self, *args):
        result = 1
        for num in args:
            result *= num
        return result
    
calc = calculator()

print(calc.add(4,6,2))
print(calc.multiply(4,6,2))
    
```

### 4. `student` class

`__init__`: takes name and grade.

Method `.greet()` → return `"Hi, I'm NAME and my grade is GRADE".`

Objects to create:

Student “Alice”, grade “B”. Call `.greet().`

Student “Bob”, grade “A”. Call `.greet().`

```python
class student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    def greet(self):
        return f"Hi, I'm {self.name} and my grade is {self.grade}"
    
alice = student("Alice", "A")
bob = student("Bob", "B")

print(alice.greet())
print(bob.greet())
```

### 5. `counter` class

`__init__`: start value = 0.

Method `.increment()` → add 1.

Method `.reset()` → set back to 0.

Objects to create:

A counter, increment 3 times.

Reset it, then increment once.

```python
class counter:
    def __init__(self, incrementor):
        self.incrementor = incrementor
        self.value = 1
    def increment(self):
        for n in range(self.incrementor):
            self.value += 1
        return self.value

inc = counter(3)

print(inc.increment())
```

### 6. `BankAccount` class

`__init__`: takes `owner` and optional `balance` (default 0).

Method `.deposit(amount)` → add to balance.

Method `.withdraw(amount)` → subtract from balance.

Method `.get_balance()` → return current balance.

Objects to create:

Account for “Alice” with £100, deposit £50, then withdraw £30.

Account for “Bob” with no starting money, deposit £200.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, deposit_amount):
        self.balance += deposit_amount
        return self.balance

    def withdraw(self, withdraw_amount):
        self.balance -= withdraw_amount
        return self.balance

    def get_balance(self):
        return self.balance


# Objects
acc_alice = BankAccount("Alice", 100)
acc_bob = BankAccount("Bob")  # balance defaults to 0

# Actions
acc_alice.deposit(50)    # 100 → 150
acc_alice.withdraw(30)   # 150 → 120
acc_bob.deposit(200)     # 0 → 200

# Results
print(acc_alice.get_balance())  # 120
print(acc_bob.get_balance())    # 200
```

### 7.`Temperature` class

`__init__`: takes celsius.

Method `.to_fahrenheit()` → convert to Fahrenheit.

Method `.to_kelvin()` → convert to Kelvin.

Objects to create:

A temperature of 25°C → convert to Fahrenheit + Kelvin.

A temperature of 0°C → convert to Fahrenheit + Kelvin.

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    
    def to_fahrenheit(self):
        return (self.celsius*1.8) + 32
    def to_kelvin(self):
        return (self.celsius+273.15)
    
object_1 = Temperature(25)
object_2 = Temperature(0)

print(f"{object_1.celsius}°C = {object_1.to_fahrenheit()}°F = {object_1.to_kelvin()}K")
print(f"{object_2.celsius}ºC = {object_2.to_fahrenheit()}ºF = {object_2.to_kelvin()}K")
```
