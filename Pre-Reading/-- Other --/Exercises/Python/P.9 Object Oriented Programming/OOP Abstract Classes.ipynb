### 1. Define an abstract class `Appliance` with abstract method `operate()`. Create `WashingMachine` and `Fridge` that implement it.

```python
from abc import ABC, abstractmethod

class Appliance(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def operate(self):
        pass

class WashingMachine(Appliance):
    def __init__(self, brand):
        super().__init__(brand)
    
    def operate(self):
        print(f"The {self.brand} washing machine is running.")

class Fridge(Appliance):
    def __init__(self, brand):
        super().__init__(brand)
    
    def operate(self):
        print(f"The {self.brand} fridge is running.")

washingmachine1 = WashingMachine("Bosch")
fridge1 = Fridge("LG")

washingmachine1.operate()
fridge1.operate()
```

### 2. Create an abstract class `Shape` with abstract method `area()`. Implement `Square` and `Circle` classes with their own `area()`.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, shape_name, shape_area):
        self.name = shape_name
        self.shape_area = shape_area
    
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self,shape_name,shape_area):
        super().__init__(shape_name, shape_area)
    
    def area(self):
        print(f"The square {self.name}'s area is {self.shape_area} centimetres cubed")

class Circle(Shape):
    def __init__(self,shape_name, shape_area):
        super().__init__(shape_name, shape_area)
    
    def area(self):
        print(f"The circle {self.name}'s area is {self.shape_area} centimetres cubed")
    
square1 = Square("square1", 90)
circle1 = Circle("circle1", 50)

square1.area()
circle1.area()
```

### 3. Create an abstract class `Employee` with abstract method `work()`. Implement `Teacher` and `Doctor` that define `work()`.

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self,name):
        self.name = name
    
    @abstractmethod
    def work(self):
        pass

class Doctor(Employee):
    def __init__(self, name):
        super().__init__(name)

    def work(self):
        print(f"{self.name} is now working")

class Student(Employee):
    def __init__(self, name):
        super().__init__(name)

    def work(self):
        print(f"{self.name} is now working")

doctor1 = Doctor("Alice")
student1 = Student("Bob")

student1.work()
doctor1.work()
```

### 4. Create an abstract class `Vehicle` with abstract method `drive()`. Implement `Car` and `Motorbike` with their own `drive()` implementations.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    def __init__(self, brand):
        super().__init__(brand)
    
    def drive(self):
        print(f"The {self.brand} car is driving")

class Motorbike(Vehicle):
    def __init__(self, brand):
        super().__init__(brand)
    
    def drive(self):
        print(f"The {self.brand} motorbike is driving")

bmw_car = Car("BMW")
honda_bike = Motorbike("Honda")

bmw_car.drive()
honda_bike.drive()
```

### 5. Create an abstract class `Animal` with abstract method `sound()`. Implement `Dog` and `Cat` classes that define `sound()`.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print(f"The dog {self.name} barks")

class Cat(Animal):
    def sound(self):
        print(f"The cat {self.name} meows")

dog1 = Dog("Alice")
cat1 = Cat("Bob")

cat1.sound()
dog1.sound()
```
