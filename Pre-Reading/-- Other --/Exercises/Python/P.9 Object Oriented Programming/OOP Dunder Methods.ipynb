### 1. Create a class `Book` with attributes `title` and `author`. Implement `__str__` so `print(book)` gives a nice message.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def __str__(self):
        return f"{self.title} by {self.author}"
```

### 2. Create a class `Team` that stores a list of players. Implement `__len__` so `len(team)` returns the number of players.

```python
class Team:
    def __init__(self, players):
        self.players = players
    def __len__(self):
        return len(self.players)

team = Team(["Alice", "Bob", "Charlie"])

print(len(team))
```

### 3. Create a class `Box` with a volume attribute. Implement `__add__` so adding two boxes gives the sum of their volumes.

```python
class Box:
    def __init__(self, volume):
        self.volume = volume
    def __add__(self, other):
        return self.volume + other.volume

box1 = Box(20)
box2 = Box(40)

print(box1+box2)
```

### 4. Create a class `Point(x, y)`. Implement `__eq__` so two points are equal if both coordinates match.

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)

point1 = Point(60,100)
point2 = Point(60,100)

print(point1 == point2)
```

### 5. Create a class `Deck` storing a list of cards. Implement `__getitem__` so you can index into the deck with `deck[0]`.

```python
class Deck:
    def __init__(self, cards):
        self.cards = cards
    def __getitem__(self, index_no):
        return self.cards[index_no]

cards = ["card1", "card2", "card3"]
deck = Deck(cards)

print(deck[1])
```
