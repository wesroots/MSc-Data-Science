### 2. Extend `Student` by adding getter and setter methods for the private `__id`. Show that direct access fails but methods work.

```python
class Student:
    def __init__(self, name, grade, id):
        self.name = name
        self._grade = grade
        self.__id = id
    def get_id(self):
        return self.__id
    def set_id(self, new_id):
        self.__id = new_id
student1 = Student("Alice", "A", "0001")

print(student1.set_id("0001"))
print(student1._Student__id) # This statement works - "print(student1.__id)" would not
```

### 3. Create a class `BankAccount` with a private attribute `__balance`. Add methods `deposit(amount)` and `get_balance()`. Show safe access.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 500
    def deposit(self, amount):
        self.__balance += amount
        return None
    def get_balance(self):
        return self.__balance

account1 = BankAccount()

account1.deposit(200)
print(f"Account 1 balance: £{account1._BankAccount__balance}")
```

### 5. Create a class `Person` with a private attribute `__age`. Use getter/setter methods to control valid ages (e.g. reject negatives).

```python
class NegativeAgeError(Exception):
            pass

class Person:
    def __init__(self, age):
        if age <0:
            raise NegativeAgeError("Age cannot be negative.")
        else:
            self.__age = age
    
    def get_age(self):
        print(self.__age)
    def set_age(self, new_age):
        if new_age <0:
            raise NegativeAgeError("Age cannot be negative.")
        else:
            self.__age = new_age

person1 = Person(23)
person1.set_age(24)
person1.get_age()
```
